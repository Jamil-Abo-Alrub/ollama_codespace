```python
import mlflow.sklearn
import pandas as pd
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

# 1. Tell FastAPI where your MLflow server is
mlflow.set_tracking_uri("http://mlflow-server-614417916386.europe-west9.run.app/#/:8080")

# 2. Load your model from the registry (use the name you gave it)
model_name = "MathScorePredictor"
model_version = 1
model = mlflow.sklearn.load_model(model_uri=f"models:/{model_name}/{model_version}")

class InputData(BaseModel):
    gender: str
    race_ethnicity: str
    parental_level_of_education: str
    lunch: str
    test_preparation_course: str

@app.post("/predict")
def predict(data: InputData):
    # Convert input to DataFrame
    df = pd.DataFrame([data.model_dump()])
    
    # Rename columns to match what the model expects
    df.columns = ["gender", "race/ethnicity", "parental level of education", "lunch", "test preparation course"]
    
    prediction = model.predict(df)
    return {"math_score": float(prediction[0])}
```
